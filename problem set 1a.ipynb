Question 3 C#
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c)
, where
s = (a+b+c)/2.
>>> triangleArea(2,2,2)
1.7320508075688772

```
int a = 2;
int b = 2;
int c = 2;

// Calculate the semi-perimeter (s) of the triangle using the formula s = (a + b + c) / 2.0

double s =(a+b+c)/2.0;
double triangleArea = Math.Sqrt(s*(s-a)*(s-b)*(s-c));  // Calculate the area of the triangle using Heron's formula
Console.WriteLine($"triangleArea({a}, {b}, {c})={triangleArea}");
```

element - 1 : 47

Expected Output:
The Even elements are:
42 56 32
The Odd elements are :
25 47

```
int[] array ={25, 47, 42, 56,32};


int[] oddArray = new int[2];        // Define two arrays to store even and odd numbers
int[] evenArray = new int[3];

int odd = 0;           // Initialize counters for even and odd numbers
int even = 0;

// Loop through each number in the original array

foreach (int num in array)
{
    if (num % 2 ==0)    // Check if the number is even
        {
           evenArray[even] = num;
           even++;
        }
    else
         {
            oddArray[odd] = num;
            odd++;
         }
}
Console.WriteLine("The even elements are: ");
for (int i = 0; i < evenArray.Length; i++)
       {
          Console.WriteLine(evenArray[i]);
       }
Console.WriteLine(" The odd elements are: ");
for(int i = 0; i < oddArray.Length; i++)
       {
          Console.WriteLine(oddArray[i]);
       }
```
